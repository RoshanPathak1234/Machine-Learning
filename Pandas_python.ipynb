## Reading files

```python
df = pd.read_csv("covid_19_data.csv")
df
```

```python
df.drop(['SNo','Last Update'] , axis=1 , inplace=True)
df.head()
```

```python
df.rename(columns={'ObservationDate' : 'Date' , 'Province/State' : 'Province' , 'Country/Region' : 'Country' }, inplace=True)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.fillna('NA')
```

```python
df.info()
```

```python
df.head(10)
```

```python
df2 = df.groupby('Country')[['Confirmed' , 'Deaths' , 'Recovered']].sum().reset_index()
df2
```

```python
df3 = df.groupby(['Country','Date'])[['Confirmed' , 'Deaths' , 'Recovered']].sum().reset_index()
df3
```

```python
df4 = df3[df3['Confirmed']>100]
df4
```
