```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('classic')
x = np.linspace(0,100,1000)
y = np.cumsum(np.random.randn(1000,5),0)
plt.plot(x,y)
plt.legend('ABCDE',ncol=2,loc='upper left')
plt.show()
```

```python
sns.set_theme()
plt.plot(x,y)
plt.legend('ABCDE' , ncol=2 , loc = 'upper left')
plt.show()
```

```python
x = pd.Series(50*np.random.randn(5000))
y = pd.Series(200*np.random.randn(5000))
z = pd.Series(100*np.random.randn(5000) + 500)
plt.hist(x , alpha=0.5,density=True)
plt.hist(y,alpha=0.5 , density=True)
plt.hist(z,alpha=0.5 , density=True)
plt.legend('xyz')
```

```python
sns.set_theme()
x = pd.Series(50*np.random.randn(5000))
y = pd.Series(200*np.random.randn(5000))
z = pd.Series(100*np.random.randn(5000)+500)
df = pd.DataFrame({'X' : x , 'Y' : y , 'Z' : z})
for col in df.columns :
    plt.hist(df[col] , alpha=0.5 , density=True)
    sns.kdeplot(df[col],fill=True)
```

```python
for col in df.columns:
    # sns.histplot(df[col])
    sns.displot(df[col])
```

```python
with sns.axes_style('white'):
    sns.jointplot(data=df,x="X",y="Y",kind='kde')
```

```python
sns.pairplot(df)
```
