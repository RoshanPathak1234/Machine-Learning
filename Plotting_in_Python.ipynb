```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
plt.plot(x,y)
```

```python
plt.scatter(x[:30],y[:30])
plt.show()
```

```python
plt.scatter(x[::10],y[::10] ,color='r')
plt.show()
```

```python
plt.plot(x,y,color='r')
plt.plot(x,np.cos(x),color='b')
```

### color properties : 
        -g : solid green 
        --c : dashed cyon 
        -.k : dashed black  
        :r : dotted red 

```python
plt.plot(x,y,'-.g')
plt.plot(x,np.cos(x),':c')
```
